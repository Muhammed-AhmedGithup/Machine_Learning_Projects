```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/ChurnData.csv')
df.head()
```

```python
df = df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip',   'callcard', 'wireless','churn']]
df['churn'] = df['churn'].astype('int')
df.head()
```

```python
x=df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip']].values
x[0:5]
```

```python
y=df['churn'].values
y[0:5]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=StandardScaler().fit(x).transform(x)
x[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression(C=0.01,solver='liblinear')
lg
```

```python
lg.fit(x_train,y_train)
```

```python
yhat=lg.predict(x_test)
```

```python
print(y_test[0:5])
print(yhat[0:5])
```
