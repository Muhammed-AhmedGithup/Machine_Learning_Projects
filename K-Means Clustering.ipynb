```python
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%204/data/Cust_Segmentation.csv")
df.head()
```

```python
df.drop('Address',axis=1,inplace=True)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
x=df.values[:,1:]
x=np.nan_to_num(x)
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(init='k-means++',n_clusters=3,n_init=12)
km.fit(x)
labels=km.labels_
print(labels)
```

```python
df["Clus_km"] = labels
df.head(5)
```
